```python
import cv2 
import numpy as np
import pywt
import difflib
import matplotlib.pyplot as plt
%matplotlib inline

def detectWatQuan(wat_embedded,image):
    imageAftWav=pywt.wavedec2(image,"db4",mode="periodization",level=3,axes=(-2, -1))
    hl3=imageAftWav[1][0]
    t2=200
    t1=115
    X2=10
    watLen=0
    s1=0
    for i in range(40):
        for j in range(40):
            if abs(hl3[i][j])>=t1+X2 and abs(hl3[i][j])<=t2-X2:
                watLen+=1
    wat_detected=np.empty(watLen,dtype=bool)
    watNo=0
    for i in range(40):
        for j in range(40):
            if abs(hl3[i][j])>=t1+X2 and abs(hl3[i][j])<=t2-X2:
                if abs(hl3[i][j])<(t1+t2)/2:
                    wat_detected[watNo]=False
                if abs(hl3[i][j])>(t1+t2)/2:
                    wat_detected[watNo]=True
                watNo+=1
    return difflib.SequenceMatcher(None,wat_embedded,wat_detected).quick_ratio()
def saltAndPepper(img,snr):
    h=img.shape[0]
    w=img.shape[1]
    img1=img.copy()
    sp=h*w   # 计算图像像素点个数
    NP=int(sp*(1-snr))   # 计算图像椒盐噪声点个数
    for i in range (NP):
        randx=np.random.randint(1,h-1)   # 生成一个 1 至 h-1 之间的随机整数
        randy=np.random.randint(1,w-1)   # 生成一个 1 至 w-1 之间的随机整数
        if np.random.random()<=0.5:   # np.random.random()生成一个 0 至 1 之间的浮点数
            img1[randx,randy]=0
        else:
            img1[randx,randy]=255
    return img1

image = cv2.imread("./image/cover.png",cv2.IMREAD_GRAYSCALE)
np.random.seed(261025)
imageAftWav=pywt.wavedec2(image,"db4",mode="periodization",level=3,axes=(-2, -1))
hl3=imageAftWav[1][0]
t2=200
t1=115
X1=20
watLen=0
for i in range(40):
    for j in range(40):
        if (hl3[i][j]>t1 and hl3[i][j]<t2) or (hl3[i][j]<-t1 and hl3[i][j]>-t2):
            watLen+=1
wat=np.random.choice(a=[False,True],size=watLen)
watNo=0
for i in range(40):
    for j in range(40):
        if (hl3[i][j]>115 and hl3[i][j]<200) and wat[watNo]==True:
            hl3[i][j]=t2-X1
            watNo+=1
        elif (hl3[i][j]>115 and hl3[i][j]<200) and wat[watNo]==False:
            hl3[i][j]=t1+X1
            watNo+=1
        elif (hl3[i][j]<-115 and hl3[i][j]>-200) and wat[watNo]==True:
            hl3[i][j]=-t2+X1
            watNo+=1
        elif (hl3[i][j]<-115 and hl3[i][j]>-200) and wat[watNo]==False:
            hl3[i][j]=-t1-X1
            watNo+=1  
            
            
imageAftIdwt=pywt.waverec2(imageAftWav,"db4",mode="periodization",axes=(-2, -1))
pic=(imageAftIdwt).astype(np.uint8)
base_ratio=detectWatQuan(wat,pic)

# blur
blurred=cv2.GaussianBlur(pic,(5,5),0)
#sharpen
sharpened=cv2.Laplacian(pic,cv2.CV_16S,5)
abs_sharpened=cv2.convertScaleAbs(sharpened)
#rotate
h,w=pic.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -90, 1)
rotated=cv2.warpAffine(pic, M, (w, h))
#crop
cropped=image
cropped[200:320,0:320]=0
cropped[0:320,200:320]=0
#compress
cv2.imwrite("./image/ReferenceImg/compressed.jpg", pic, [cv2.IMWRITE_JPEG_QUALITY, 5])
compressed=cv2.imread("./image/ReferenceImg/compressed.jpg",cv2.IMREAD_GRAYSCALE)
# salt and pepper
saltAndPep=saltAndPepper(pic,0.8)


#cv2.imwrite("./image/ReferenceImg/blurred.jpg",blurred,[cv2.IMWRITE_JPEG_QUALITY, 100])
#cv2.imwrite("./image/ReferenceImg/sharpened.jpg",abs_sharpened,[cv2.IMWRITE_JPEG_QUALITY, 100])
#cv2.imwrite("./image/ReferenceImg/rotated.jpg",rotated,[cv2.IMWRITE_JPEG_QUALITY, 100])
#cv2.imwrite("./image/ReferenceImg/cropped.jpg",cropped,[cv2.IMWRITE_JPEG_QUALITY, 100])
#cv2.imwrite("./image/ReferenceImg/compressed.jpg",compressed,[cv2.IMWRITE_JPEG_QUALITY, 100])
#cv2.imwrite("./image/ReferenceImg/saltAndPep.jpg",saltAndPep,[cv2.IMWRITE_JPEG_QUALITY, 100])
#cv2.imwrite("./image/ReferenceImg/beforeAttack.jpg",pic,[cv2.IMWRITE_JPEG_QUALITY, 100])


blurred_ratio=detectWatQuan(wat,blurred)
sharpened_ratio=detectWatQuan(wat,abs_sharpened)
rotated_ratio=detectWatQuan(wat,rotated)
cropped_ratio=detectWatQuan(wat,cropped)
compressed_ratio=detectWatQuan(wat,compressed)
saltAndPep_ratio=detectWatQuan(wat,saltAndPep)
print(cropped_ratio/base_ratio)
print(compressed_ratio/base_ratio)
print(sharpened_ratio/base_ratio)
print(saltAndPep_ratio/base_ratio)
print(base_ratio)


#cv2.waitKey()
```
